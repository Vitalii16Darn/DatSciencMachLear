```python
#пошук відповідного середнього значення
def related_mean(df, column, related):
    related_means = dict()
    for _, row in df.iterrows():
        if row[column] is None or row[related] is None:
            continue
        if row[related] not in related_means:
            related_means[row[related]] = np.array([])

        related_means[row[related]] = np.append(related_means[row[related]], [float(row[column])])
    for related in related_means.keys():
        related_means[related] = np.mean(related_means[related]).round(0)

    return related_means

normalized_losses_mean = related_mean(with_nones, 'normalized-losses', 'symboling')
horsepower_mean = related_mean(with_nones, 'horsepower', 'curb-weight')
display(normalized_losses_mean)
display(horsepower_mean)
```

```python
def replace(df, c_to_replace, ref_column, ref_data, default_value=None):
    for index, row in df.iterrows():
        if row[c_to_replace] == None:
            if df[ref_column][index] in ref_data.keys():
                df[c_to_replace][index] = ref_data[df[ref_column][index]]
                continue
            df[c_to_replace][index] = default_value
replace(with_nones, 'normalized-losses', 'symboling', normalized_losses_mean, np.mean(list(normalized_losses_mean.values())).round(1))
replace(with_nones, 'horsepower', 'curb-weight', horsepower_mean, np.mean(list(horsepower_mean.values())).round(1))
```

```python
price_mean = related_mean(with_nones, 'price', 'horsepower')
bore_mean = related_mean(with_nones, 'bore', 'horsepower')
stroke_mean = related_mean(with_nones, 'stroke', 'horsepower')
peak_rpm_mean = related_mean(with_nones, 'peak-rpm', 'horsepower')


display(price_mean)
display(bore_mean)
display(stroke_mean)
display(peak_rpm_mean)

replace(with_nones, 'price', 'horsepower', price_mean, np.mean(list(price_mean.values())).round(1))
replace(with_nones, 'bore', 'horsepower', bore_mean, np.mean(list(bore_mean.values())).round(1))
replace(with_nones, 'stroke', 'horsepower', stroke_mean, np.mean(list(stroke_mean.values())).round(1))
replace(with_nones, 'peak-rpm', 'horsepower', peak_rpm_mean, np.mean(list(peak_rpm_mean.values())).round(1))
```

```python
from collections import Counter

#залежність між кількістю дверей та типом кузову
def related(df, column, related):
    relateds = dict()
    for _, row in df.iterrows():
        if row[column] is None or row[related] is None:
            continue
        if row[related] not in relateds:
            relateds[row[related]] = np.array([])

        relateds[row[related]] = np.append(relateds[row[related]], [row[column]])
    for related in relateds.keys():
        relateds[related] = dict(Counter(relateds[related]))

    for key, value in relateds.items():
        relateds[key] = list(relateds[key].keys())[0]

    return relateds
num_of_doors_mean = related(with_nones, 'num-of-doors', 'body-style')
display(num_of_doors_mean)

replace(with_nones, 'num-of-doors', 'body-style', num_of_doors_mean)
display(with_nones.isnull().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

transformed = with_nones
transformed['symboling'] = MinMaxScaler().fit_transform(transformed['symboling'].values.reshape((-1, 1)))
transformed['make'] = LabelEncoder().fit_transform(transformed['make'].values)
transformed['fuel-type'] = LabelEncoder().fit_transform(transformed['fuel-type'].values)
transformed['aspiration'] = LabelEncoder().fit_transform(transformed['aspiration'].values)
transformed['num-of-doors'] = LabelEncoder().fit_transform(transformed['num-of-doors'].values)
transformed['body-style'] = LabelEncoder().fit_transform(transformed['body-style'].values)
transformed['drive-wheels'] = LabelEncoder().fit_transform(transformed['drive-wheels'].values)
transformed['engine-location'] = LabelEncoder().fit_transform(transformed['engine-location'].values)
transformed['engine-type'] = LabelEncoder().fit_transform(transformed['engine-type'].values)
transformed['fuel-system'] = LabelEncoder().fit_transform(transformed['fuel-system'].values)
transformed['num-of-cylinders'] = LabelEncoder().fit_transform(transformed['num-of-cylinders'].values)
transformed['price'] = transformed['price'].apply(lambda x: float(x))
transformed['normalized-losses'] = transformed['normalized-losses'].apply(lambda x: float(x))

display(transformed)

working = transformed
working_with_noprice = working.drop(["price"], axis=1)
working_price = working["price"]
```
